```python
import cnotebook
import oecluster
import oepandas as oepd
```

# Molecular Clustering with OECluster

```python
df = oepd.read_molecule_csv("assets/Bender.csv", molecule_columns="Molecule")

df.head()
```

```python
morgan_dmat = oecluster.pdist(
    df.Molecule.tolist(),
    "fingerprint",
    fp_type="morgan",
    metric="jaccard",
    num_threads=4
)
```

```python
butina_result = oecluster.butina(morgan_dmat, threshold=0.6)

# .clusters keeps the representative-first member lists; .labels is a
# length-n assignment array suitable for a DataFrame column.
butina_result.clusters
```
